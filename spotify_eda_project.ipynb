```python
import pandas as pd
import zipfile
from google.colab import files

# Upload the ZIP file
uploaded = files.upload()

# Unzip the file
with zipfile.ZipFile("dataset.csv.zip", 'r') as zip_ref:
    zip_ref.extractall()

# Now read the CSV file (usually named tracks.csv inside the zip)
df = pd.read_csv("dataset.csv")

# Show first 5 rows
df.head()
```
