```python
import pandas as pd
import zipfile
from google.colab import files

# Upload the ZIP file
uploaded = files.upload()

# Unzip the file
with zipfile.ZipFile("dataset.csv.zip", 'r') as zip_ref:
    zip_ref.extractall()

# Now read the CSV file (usually named tracks.csv inside the zip)
df = pd.read_csv("dataset.csv")

# Show first 5 rows
df.head()
```

```python
# Check the column names and data types
df.info()
```

```python
# Quick summary statistics
df.describe()
```

```python
# Check if any columns have missing values
df.isnull().sum()
```

```python
# View unique genres
df['track_genre'].unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_genres = df['track_genre'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette="viridis")
plt.title("Top 10 Genres on Spotify")
plt.xlabel("Number of Songs")
plt.ylabel("track_genre")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['popularity'], bins=30, kde=True, color='orange')
plt.title('Distribution of Song Popularity')
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='danceability', y='energy', data=df, alpha=0.5)
plt.title('Danceability vs Energy')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.show()
```

```python
plt.figure(figsize=(12,8))
numeric_cols = df.select_dtypes(include='number')
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Numeric Features')
plt.show()
```

## 📌 Key Insights

- Pop music is the most dominant genre in the dataset.
- Most songs have popularity below 60 — only a few are viral hits.
- Danceability and energy show a positive relationship — party songs are both energetic and danceable.
- Loudness and energy are strongly correlated, as expected.
